```
///<summary>
/// Calculates the square of a number
/// </summary>
/// <param name="number"> input number.</param>
/// <returns>square of a number.</returns>
static int Square(int number)
{
    return number * number;
}
```

```
/// <summary> Calculates the average of a number</summary>
/// <param name="score1"> first number.</param>
/// <param name="score2"> second number.</param> 
/// <param name="score3"> third number.</param>
/// <returns> average of thiry numbers.</return>
static double CalculateAverage(int score1, int score2, int score3)
{
    return (score1 + score2 + score3) / 3.0;
}
```
