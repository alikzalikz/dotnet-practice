```
static string EvenOrOdd(int number)
{
    if (number % 2 == 0)
    {
        return "Even";
    }
    else
    {
        return "Odd";
    }
}

static void RunEvenOrOdd()
{
    for (int num = 1; num <= 10; num++)
    {
        Console.WriteLine($"{num}: {EvenOrOdd(num)}");
    }

    Console.WriteLine();
} RunEvenOrOdd();
```

```
static string TemperatureStatus(int temperature)
{
    if (temperature < 0)
    {
        return "freezing";
    }
    else if(0<=temperature && temperature <= 15)
    {
       return "cold"; 
    }
    else if(16<=temperature && temperature <= 25)
    {
        return "warm";
    }
    else
    {
        return "hot";
    }
}
static void RunTemperatureStatus()
{
    for (int i= -5; i<=35; i+=5)
    {
        
        Console.WriteLine($"{i}:{TemperatureStatus(i)}");

    }
}RunTemperatureStatus();
```

```
static string GetMedal(int score)
{
    if(score<=100 && score >= 90)
    {
        return "Gold";
    }
    else if (score <=89 && score >=80 )
    {
        return "silver";
    }
    else if (score <=79 && score >= 70)
    {
        return "boronz";
    }
    else
    {
        return "no medal";
    }


}
static void RunMedals()
{
    for (int i =50; i<=100; i+=10)
    {
        Console.WriteLine($"{i}:{GetMedal(i)}");
    }
}
RunMedals();


```
