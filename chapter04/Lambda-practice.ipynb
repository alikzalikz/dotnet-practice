```
static string GetSeason(int month) => month switch
{
    1 or 2 or 3 => "winter",
    4 or 5 or 6 => "spring",
    7 or 8 or 9 => "summer",
   10 or 11 or 12 => "autumn",
    _ => "invalid"
} ;
Console.WriteLine(GetSeason(5));
```

```
static string GetGrade(int score) => score switch
{
    >= 90 and <= 100 => "A",
    >= 80 and < 90   => "B",
    >= 70 and < 80   => "C",
    >= 60 and < 70   => "D",
    >= 0  and < 60   => "F",
    _ => "Invalid"
};
Console.WriteLine($"{getgrade(65)}");
```
