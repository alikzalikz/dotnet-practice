```

using System.Globalization;
decimal price = 1250.50m;
CultureInfo.CurrentCulture=CultureInfo.GetCultureInfo("en-US");
Console.WriteLine($"{price:C}");
```

```
using System.Globalization;
decimal price = 999.99m;
CultureInfo.CurrentCulture=CultureInfo.GetCultureInfo("en-US");
Console.WriteLine($"{price:C}");

CultureInfo.CurrentCulture=CultureInfo.GetCultureInfo("fr-FR");
Console.WriteLine($"{price:C}");

CultureInfo.CurrentCulture=CultureInfo.GetCultureInfo("de-DE");
Console.WriteLine($"{price:C}");


```

```
using System.Globalization;

static void ShowPrice(decimal price, string code)
{
    CultureInfo.CurrentCulture = CultureInfo.GetCultureInfo(code);

    Console.WriteLine($"{price:C}");
}

ShowPrice(10055m, "fr-FR");
ShowPrice(10055m, "en-US");
ShowPrice(10055m, "de-DE");
```

```
using System.Globalization;

static void ShowDate(DateTime date,string code){
    CultureInfo.CurrentCulture=CultureInfo.GetCultureInfo(code);
    Console.WriteLine($"{date:D}");
}
ShowDate(DateTime.Now, "en-US");
ShowDate(DateTime.Now, "fr-FR");
```
